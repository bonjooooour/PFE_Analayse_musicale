```python
import os
# Spécifiez le chemin du dossier racine
dossier_racine = r'C:\Users\lucie\Documents\Programmation\McGill-Billboard'

# Liste pour stocker tous les fichiers trouvés
tous_les_fichiers = []

# Parcourir récursivement l'arborescence
for dossier, sous_dossiers, fichiers in os.walk(dossier_racine):
    for fichier in fichiers:
        chemin_fichier = os.path.join(dossier, fichier)
        tous_les_fichiers.append(chemin_fichier)

```

```python
lignes_list = []

for fichier in tous_les_fichiers:
    with open(fichier, 'r') as fichier_ouvert:
        lignes = fichier_ouvert.readlines()[5:]  # Ignorer les 5 premières lignes
        # Faites quelque chose avec le reste des lignes du fichier
        # Par exemple, afficher les 5 lignes suivantes de chaque fichier
        lignes_list.extend(lignes)
```

```python
import pandas as pd
from io import StringIO

donnees_fichier = StringIO("\n".join(lignes_list))

# Créer le DataFrame
colonnes = ["Temps", "Label"]
df = pd.read_csv(donnees_fichier, sep='\t', names=colonnes)

# Afficher les premières lignes du DataFrame
df.head()
```

```python
labels = df['Label'].str.split(',')
chords_seq = []

for i in range(len(labels)):
    if isinstance(labels[i], list):
        if len(labels[i]) == 1:
            chords_seq.append('nan')
        elif len(labels[i]) == 2:
            chords_seq.append(labels[i][0])
        else:
            chords_seq.append(labels[i][2])
    else:
        # Gérer le cas où la valeur n'est pas une liste (peut être un NaN)
        chords_seq.append('nan')
```

```python
liste_chords = []
temps = df['Temps']
liste_temps = []

for i in range(len(chords_seq)):
    if('|' in chords_seq[i]):
        val = chords_seq[i].split('|')
        val = val[1:-1]
        for elem in val:
            elem = elem.replace(':','')
            elem = elem.replace(' ','')
            liste_chords.append(elem)
            liste_temps.append(temps[i])
    else:
          liste_chords.append(chords_seq[i])
          liste_temps.append(temps[i])
```

```python
len(liste_temps),len(liste_chords)
```

```python
df_final = pd.DataFrame()
df_final['start'] = liste_temps
df_final['chords'] = liste_chords
df_final.head()
```

```python
chemin_fichier_excel = r'Chords_analysisexcel.xlsx'

# Sauvegarder le DataFrame dans le fichier Excel
df_final.to_excel(chemin_fichier_excel, index=False)

print(f"Le DataFrame a été sauvegardé dans {chemin_fichier_excel}")
```
